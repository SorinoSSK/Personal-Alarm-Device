```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "cleaned_exported_data1"
```

```python
df = pd.read_csv('../DataFile/' + filename + '.csv')
if 'level_0' in df.columns:
    df.drop(columns=['level_0'], inplace=True) # drop columns with NaN values and gibberish
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True) # drop columns with NaN values and gibberish
if 'Unnamed: 0.1' in df.columns:
    df.drop(columns=['Unnamed: 0.1'], inplace=True) # drop columns with NaN values and gibberish
df = df.reset_index()
display(df)
```

```python
df = df.drop(df.index[142820:df.index[-1]+1])
display(df)
```

```python
import os

os.chdir("../DataFile/")
print(os.getcwd())

count = 0
for path in os.scandir(os.getcwd()):
    if path.is_file() and path.name.endswith('.csv'):
        if 'cleaned_exported_data' in path.name:
            count+=1
count+=1
newPath = '../DataFile/cleaned_exported_data' + str(count) + '.csv'
df.to_csv(newPath)
```
