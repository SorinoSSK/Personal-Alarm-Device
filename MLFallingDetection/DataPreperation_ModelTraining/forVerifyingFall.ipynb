```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "cleaned_exported_data2"
```

```python
df = pd.read_csv('../DataFile/' + filename + '.csv')
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True) # drop columns with NaN values and gibberish
if 'level_0' in df.columns:
    df.drop(columns=['level_0'], inplace=True) # drop columns with NaN values and gibberish
if 'index' in df.columns:
    df.drop(columns=['index'], inplace=True) # drop columns with NaN values and gibberish
df = df.reset_index()
display(df)
```

```python
FallData = pd.DataFrame(df[df['label'] == 1])
display(FallData)
```

```python
# oldIndex = FallData.iloc[0]['index']
oldCompIndex = FallData.iloc[0]['index']
oldLstIndex = 0
SkyIsFalling = []
reachEnd = False
for lstindex, (index, data) in enumerate(FallData.iterrows()):
    if index - oldCompIndex > 1:
        SkyIsFalling.append(FallData.iloc[oldLstIndex:lstindex])
        oldLstIndex = lstindex
    oldCompIndex = index

print("Total samples:", SkyIsFalling[0])
```

```python
# fig = plt.figure(len(SkyIsFalling))
fig, axs = plt.subplots(len(SkyIsFalling), figsize=(8, 100), tight_layout=True)
# fig, axs = plt.subplots(3)
for index, value in enumerate(SkyIsFalling):
    axs[index].plot(value['index'], value['Norm'])
    titleString = str(index+1) + ": " + str(value['index'].iloc[0]) + " - " + str(value['index'].iloc[-1])
    axs[index].set_title(titleString)
    display(value)
```
