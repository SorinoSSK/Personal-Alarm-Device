```python
import pandas as pd
import numpy as np
```

```python
filename1 = "FallDataExport2"
filename2 = "exported_data_4"
```

```python
df1 = pd.read_csv('../DataFile/' + filename1 + '.csv')
```

```python
if 'level_0' in df1.columns:
    df1.drop(columns=['level_0'], inplace=True) # drop columns with NaN values and gibberish
if 'Unnamed: 0' in df1.columns:
    df1.drop(columns=['Unnamed: 0'], inplace=True) # drop columns with NaN values and gibberish
df1.head()
```

```python
if not 'Norm' in df1.columns:
    df1['Norm'] = np.sqrt(df1['accX']**2+df1['accY']**2+df1['accZ']**2)
    df1 = df1.sort_values(by=['timestamp'])
df1 = df1.reset_index()
df1 = df1.reset_index()
# df1['label'] = 0
df1.head()
```

```python
df1 = df1.drop(['index'], axis=1)
df1.head()
```

```python
# df0 = pd.read_csv('combined_export_v2.csv')
df0 = pd.read_csv('../DataFile/' + filename2 + '.csv')
if 'level_0' in df0.columns:
    df0.drop(columns=['level_0'], inplace=True) # drop columns with NaN values and gibberish
if 'index' in df0.columns:
    df0.drop(columns=['index'], inplace=True) # drop columns with NaN values and gibberish
if 'Unnamed: 0' in df0.columns:
    df0.drop(columns=['Unnamed: 0'], inplace=True) # drop columns with NaN values and gibberish
if 'Unnamed: 0.1' in df0.columns:
    df0.drop(columns=['Unnamed: 0.1'], inplace=True) # drop columns with NaN values and gibberish
if not 'Norm' in df0.columns:
    df0['Norm'] = np.sqrt(df0['accX']**2+df0['accY']**2+df0['accZ']**2)
    df0 = df0.sort_values(by=['timestamp'])
# df0 = df0.reset_index()
# df0 = df0.reset_index()
display(df0)
df3 = pd.concat([df1,df0])
if 'level_0' in df3.columns:
    df3.drop(columns=['level_0'], inplace=True) # drop columns with NaN values and gibberish
if 'index' in df3.columns:
    df3.drop(columns=['index'], inplace=True) # drop columns with NaN values and gibberish
df3 = df3.reset_index()
df3.drop(columns=['index'], inplace=True)

display(df3)
```

```python
import os

os.chdir("../DataFile/")
print(os.getcwd())

count = 0
for path in os.scandir(os.getcwd()):
    if path.is_file() and path.name.endswith('.csv'):
        if 'FallDataExport' in path.name:
            count+=1
count+=1
newPath = '../DataFile/FallDataExport' + str(count) + '.csv'
df3.to_csv(newPath)
```
