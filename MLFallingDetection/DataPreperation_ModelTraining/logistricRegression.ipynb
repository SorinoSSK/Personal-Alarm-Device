```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,log_loss
```

```python
filename = "FallDataExport3"
```

```python
windowFrame = 10
df = pd.read_csv('../DataFile/' + filename + '.csv')
df.drop(columns=['Unnamed: 0','timestamp', 'accX', 'accY', 'accZ'],inplace=True)
display(df)

col = ["Bias"]
col += (["col"+str(x) for x in range(windowFrame)])
col.append("label")
print(col)

newDict = []
dataPos, itrIndex, tempSum = 0, 0, 0
while dataPos+10 < len(df):
    labelCount = 0
    tempDict = {}
    dataPos = 10+5*itrIndex
    tempDf = df.iloc[0+5*itrIndex:dataPos]
    tempDict = {col[i+1]: tempDf.iloc[i, 0] for i in range(windowFrame)}
    tempDict["Bias"] = 1
    tempSum = tempDf.iloc[:, 1].sum()
    if tempSum >= 5:
        tempDict[col[-1]] = 1
    else:
        tempDict[col[-1]] = 0
    newDict.append(tempDict)
    itrIndex += 1
newDf = pd.DataFrame(newDict, columns=col)
display(newDf)
```

```python
X = newDf[col[:-1]].to_numpy()
Y = newDf['label'].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
loss = log_loss(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```
